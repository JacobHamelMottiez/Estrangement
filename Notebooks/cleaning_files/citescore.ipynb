```python
import pybliometrics
from pybliometrics.scopus import SerialSearch
pybliometrics.scopus.init()

cited_journals = pd.read_csv("C:/Users/jacob/OneDrive - Université Laval/biophilo/Data/ISSN_test.csv")
cited_journals = cited_journals.dropna(subset=['ISSN']).reset_index(drop=True)
cited_journals


```

```python
import pandas as pd
from pybliometrics.scopus import SerialTitle

final_df = pd.DataFrame()

for ISSN in cited_journals['ISSN'].dropna().unique():  # Ensure ISSN values are valid
    try:
        # Retrieve source title and 'n' value safely
        source_data = cited_journals.loc[cited_journals['ISSN'] == ISSN, ['sourcetitle', 'n']]
        if not source_data.empty:
            source_title = source_data['sourcetitle'].iloc[0]  # Get first matching value
            n_value = source_data['n'].iloc[0]  # Get the corresponding 'n' value
        else:
            source_title = None  # Handle missing source titles
            n_value = None  # Handle missing 'n' values
        
        # Fetch citation metrics
        metrics = SerialTitle(ISSN, view='CITESCORE')
        
        # Convert citescoreyearinfolist to DataFrame
        df = pd.DataFrame(metrics.citescoreyearinfolist)  
        
        # Ensure 'citescore' column exists before calculating statistics
        if 'citescore' in df.columns:
            mean_citescore = df['citescore'].mean()
            median_citescore = df['citescore'].median()
        else:
            mean_citescore = None
            median_citescore = None

        # Create a DataFrame with ISSN, mean, median citescore, and 'n'
        info = pd.DataFrame({
            'sourcetitle': [source_title],  # Store single value as a list
            'ISSN': [ISSN],
            'n': [n_value],  # Add the 'n' column
            'median_citescore': [median_citescore],
            'mean_citescore': [mean_citescore]
        })

        # Concatenate the DataFrame
        final_df = pd.concat([final_df, info], ignore_index=True)
    
    except Exception as e:
        print(f"Error retrieving data for ISSN {ISSN}: {e}")

print(final_df)
```

```python
final_df.to_csv("C:/Users/jacob/OneDrive - Université Laval/biophilo/Data/ISSN_median_mean.csv")
final_df
```

```python
# Calculate 'weighted_citescore' after the loop
final_df['weighted_citescore'] = final_df['n'] * final_df['mean_citescore']

# Filter rows where weighted_citescore > 500
filtered_df = final_df[final_df['weighted_citescore'] > 3000]

print(filtered_df)
filtered_df.to_csv("C:/Users/jacob/OneDrive - Université Laval/biophilo/Data/ISSN_weight_3000.csv")
```
