```python
import pandas as pd
import re
import numpy as np
import pybliometrics
from pybliometrics.scopus import ScopusSearch


doi_data = pd.read_csv("C:\\Users\\jacob\\OneDrive - Université Laval\\biophilo\\Data\\PHILPAPERS_PHILO_DATA\\philpapers_articles_classified.csv")

# FILTER PROBLEMATIC DOI --------------------------------------------------

# Create a `doi_query` column
doi_data['doi_query'] = doi_data['DOI'].apply(lambda x: f"DOI({x})")

# Filter out rows with problematic patterns
doi_data = doi_data[~doi_data['doi_query'].str.contains(r'\(.*[()]+.*\)|"', regex=True)]

# Additional filtering
test = doi_data[doi_data['doi_query'] == 'DOI(10.1016/j.red.2019.04.004)']

# FUNCTION TO GET GOOD FORMAT FOR QUERY -----------------------------------
def collapse_in_chunks(data, chunk_size=25):
    """
    Split the data into chunks of `chunk_size` rows
    and collapse each chunk into a single string.
    """
    # Split the data into chunks
    chunks = np.array_split(data['doi_query'].tolist(), 
                            np.ceil(len(data) / chunk_size).astype(int))
    
    # Collapse each chunk into a single string
    collapsed_strings = [' or '.join(chunk) for chunk in chunks]
    
    # Return a DataFrame with each chunk as a row
    return pd.DataFrame({'chunk': collapsed_strings})

result = collapse_in_chunks(doi_data, chunk_size=25)
result

```

# Articles' data

```python
result_list = []  # Initialize an empty list to store results

# Assuming result.chunk is an iterable (e.g., a list of queries or DOIs)
for query_item in result.chunk:
    pybliometrics.scopus.init()
    query = f"{query_item}"  # Use f-string for cleaner code
    print(query)
    s = ScopusSearch(query, verbose=True, subscriber=True, view="COMPLETE")
    df = pd.DataFrame(s.results)  # Add the results of each query to result_list
    result_list.append(df)  # Append each df to result_list
    

final_df = pd.concat(result_list, ignore_index=True)
final_df.to_csv("C:\\Users\\jacob\\OneDrive - Université Laval\\biophilo\\Data\\PHILPAPERS_PHILO_DATA\\classified_philpapers_data_pyblio.csv")
```
