```python

import pandas as pd
from cosmograph import cosmo

PATH = "C:/Users/jacob/OneDrive - Université Laval/biophilo/Data/pybiblio/SPECIALIZED PHILOSOPHY OF BIOLOGY/"

points = pd.read_csv(PATH + "nodes_data_cg.csv", dtype={'cited_year' : str})
links = pd.read_csv(PATH + "edges_data_cg.csv")

points = points.dropna(subset=['cited_year'])
points
```

```python
from datetime import datetime, timezone
points["date"] = points.cited_year 
points['date'] = pd.to_datetime(points['date'], errors='coerce')
points
```

```python
widget = cosmo(
  points=points,
  links=links,
  point_id_by='id',
  link_source_by='source',
  link_target_by='target',
  point_color_by='category',
  point_label_by='label',
  point_size_by='value',
  curved_links=True,
  point_size_scale=2,
  show_labels=False,  
  point_timeline_by='date',
  pixel_ratio = 15
  
)

widget

```

```python
from ipywidgets.embed import embed_minimal_html
embed_minimal_html(f'{PATH}\\test.html', views = [widget],title='test export',drop_defaults = False)
```
